# COVID 19 Bed Occupancy Model

## Problem Statement

    With the Onset of COVID 19 Pandemic , It is essential for a hospital to estimate when the maximum bed occupancy will be reached. This could be impacted by several parameters like below.
    1. Number of Admissions per day
    2. Number of Recoveries per day
    3. Total Number of Beds available

## Significance of Model

    With the above factors the saturation time for maximum bed occupancy can be calculated by running simulation keeping all the above factors as variables for n days. We can adjust the timelines (number of days to run simulation) based on the requirements. The model is built to accept the following  Input Parameters and Output Parameters
    
   ### Inputs to the Model : 
        
         1. Minimum and Maximum Number of Admissions per day - This defines a range of admissions of patients into hospital including both COVID and NON COVID patients.
         2. Minimum and Maximum Number of Discharges per day - This defines a range of patients who are discharged from hospital including both COVID and NON COVID patients.
         3. Number of days  - This defines number of days the results need to be simulated
         4. Number of Available Beds - This input marks baseline number of beds available for COVID and NON COVID patients.

   ### Outputs from the Model : 
    
        1. Total Free Beds for COVID and NON COVID
        2. Cumulative Number of COVID and NON COVID patients

### Importing Required Libraries

```python
# Configure Jupyter so figures appear in the notebook
%matplotlib inline

# Configure Jupyter to display the assigned value after an assignment
%config InteractiveShell.ast_node_interactivity='last_expr_or_assign'

# import functions from the modsim library
from modsim import *
import random
import datetime
```

### Discrete Event # 1 : 

    Patient Triage Function to classify the COVID and NON COVID Patients

```python
def patient_triage(state_patient,i,date):
    state_patient.patientid = i
    covid_choice = ['Y','N']
    state_patient.covid_diagnosed = random.choice(covid_choice)
    if(state_patient.covid_diagnosed=='Y'):
        state_patient.covid_count = state_patient.covid_count+1
        state_patient.tot_covid_count = state_patient.tot_covid_count+1
    else:
        state_patient.non_covid_count = state_patient.non_covid_count+1
        state_patient.tot_non_covid_count = state_patient.tot_non_covid_count+1
    state_patient.admit_date = date
    return state_patient
```

### Discrete Event # 2 : 

    Patient Allocate Function to assign the COVID and NON COVID Beds to patients

```python
def patient_allocate(state_beds):
    if(state_patient.covid_diagnosed == 'Y'):
        state_beds.no_of_covid_beds = state_beds.no_of_covid_beds + 1
        state_beds.no_of_free_covid_beds = state_beds.no_of_free_covid_beds - 1
    else:
        state_beds.no_of_non_covid_beds = state_beds.no_of_non_covid_beds + 1
        state_beds.no_of_free_non_covid_beds = state_beds.no_of_free_non_covid_beds - 1
    return state_beds
```

### Discrete Event # 3 : 

    Patient De-allocate Function to remove the assigneed COVID and NON COVID Beds

```python
def patient_deallocate(state_beds):
    state_patient.patientid = i
    covid_choice = ['Y','N']
    state_patient.covid_diagnosed = random.choice(covid_choice)
    if(state_patient.covid_diagnosed == 'Y'):
        state_beds.no_of_covid_beds = state_beds.no_of_covid_beds - 1
        state_beds.no_of_free_covid_beds = state_beds.no_of_free_covid_beds + 1
    else:
        state_beds.no_of_non_covid_beds = state_beds.no_of_non_covid_beds - 1
        state_beds.no_of_free_non_covid_beds = state_beds.no_of_free_non_covid_beds + 1
    return state_beds
```

### Discrete Event # 4 : 

    Patient Checkout function to assign discharge dates

```python
def patient_checkout(state_patient,i,date):
    state_patient.patientid = i
    state_patient.discharge_date = date
    return state_patient
```

### Simulation Definition : 

1. Simulation 1 - to Triage and Admit the patient into the hospital
2. Simulation 2 - to Checkout the patient out from the hospital

```python
def run_simulation_triage_admit(state_patient,state_beds,i,date):
    state_patient_upd = patient_triage(state_patient,i,date)
    state_beds_upd = patient_allocate(state_beds)
    return state_patient_upd,state_beds_upd  
def run_simulation_discharge(state_patient,state_beds,i,date):
    state_patient_upd = patient_checkout(state_patient,i,date)
    state_beds_upd = patient_deallocate(state_beds)
    return state_patient_upd,state_beds_upd  
```

### Model Execution :

Steps used to execute the model
1. Create States for Patient and Beds
2. Retrieve Number of days and create date list
3. Create Timeseries to store results and active counts
4. Call the below simulations for each date and for each patient
    1. Triage and Allocate
    2. Discharge 

```python
def execute_model(min_admit_patients,max_admit_patients,min_discharge_patients,max_discharge_patients,no_of_days,tot_covid_beds,tot_non_covid_beds):
    state_beds = State(no_of_covid_beds=0,max_covid_beds=tot_covid_beds,no_of_non_covid_beds=0,max_non_covid_beds=tot_non_covid_beds,no_of_free_covid_beds=tot_covid_beds,no_of_free_non_covid_beds=tot_non_covid_beds)
    state_patient = State(patientid=0,covid_diagnosed='N',admit_date='2020-01-07',discharge_date='2020-01-07',covid_count=0,non_covid_count=0,tot_covid_count=0,tot_non_covid_count=0)

    start_date = datetime.date(2020, 7, 1)

    date_list = []

    i=0
    while(i<=no_of_days):
        date_list.append(start_date + datetime.timedelta(days=i))
        i=i+1

    i=1
    results_non_covid = TimeSeries()
    results_covid = TimeSeries()
    covid_active = TimeSeries()
    non_covid_active = TimeSeries()
    
    results_non_covid[0]=tot_non_covid_beds
    results_covid[0]=tot_covid_beds
    covid_active[0]=0
    non_covid_active[0]=0

    for date in date_list:
        no_of_patients_admit=random.randrange(min_admit_patients, max_admit_patients, 2) 
        no_of_patients_discharge=random.randrange(min_discharge_patients,max_discharge_patients, 2) 
        state_patient.covid_count = 0 
        state_patient.non_covid_count = 0 
        patient_array_admit = linspace(1, no_of_patients_admit, no_of_patients_admit, dtype=int)
        patient_array_discharge = linspace(1, no_of_patients_discharge, no_of_patients_discharge, dtype=int)
        for patient in patient_array_admit:
            state_patient_upd,state_beds_upd = run_simulation_triage_admit(state_patient,state_beds,patient,date)
        for patient in patient_array_discharge:
            state_patient_upd,state_beds_upd = run_simulation_discharge(state_patient,state_beds,patient,date)
        if(state_beds_upd.no_of_free_covid_beds>=0):
            results_covid[i]=state_beds_upd.no_of_free_covid_beds
        if(state_beds_upd.no_of_free_non_covid_beds>=0):
            results_non_covid[i]=state_beds_upd.no_of_free_non_covid_beds
        covid_active[i]=state_patient.tot_covid_count
        non_covid_active[i]=state_patient.tot_non_covid_count
        i=i+1
    return results_covid,results_non_covid,covid_active,non_covid_active
```

### Execution # 1:

Execute the Model with the Parameters below and capture/plot the output

1. Admission Rate per day : 10-100 patients
2. Discharge Rate per day : 5-10 patients
3. Number of Days to Run simulation : 180 days
4. Number of COVID beds available : 1000
5. Number of Non COVID beds available : 500

```python
#With Admission rate 10 - 100 patients per day and recovery of 5 - 10 Patients

results_covid,results_non_covid,covid_active,non_covid_active=execute_model(10,100,5,10,180,1000,500)
plot(results_covid,label='Covid Beds')
plot(results_non_covid,label='Non Covid Beds')
decorate(xlabel='Time (days)',
         ylabel='Number of Free Beds')
```

```python
plot(covid_active,label='Active Covid Count')
plot(non_covid_active,label='Active Non Covid Count')
decorate(xlabel='Time (days)',
         ylabel='Covid Count')
```
