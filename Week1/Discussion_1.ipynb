```python
import random
from modsim import *
```

```python
l = ['H','T']
game = State(h=0, t=0,h_amt=0,t_amt=0)
```

```python
def play(state,h_win,t_win):
    out=random.choice(l)
    if(out=='H'):
        state.h += 1
        state.h_amt += h_win
    else:
        state.t += 1
        state.t_amt += t_win

def model_sim(state,num_of_toss):
    heads = TimeSeries()
    tails = TimeSeries()
    for i in range(num_of_toss):
        play(state,10,5)
        heads[i]=state.h_amt
        tails[i]=state.t_amt
    plot(heads,label="Heads")    
    plot(tails,label="Tails")    
```

```python
model_sim(game,1000)
decorate(title='Coin Toss Game',
             xlabel='No. of Tosses', 
             ylabel='Amount of Winnings')
```
